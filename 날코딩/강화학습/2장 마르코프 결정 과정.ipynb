#### 주요 가정 : Markov property
- 다음 상태는 현재 상태에만 의존함

- 모든 t, t+1, ..., T 시점에 대해서 고려해볼것임.
- 먼저 특정 t 시점에 대해서 고려해보자.
- Markov property에 의해 t+1 시점은 t 시점에 의해서만 영향을 받음.
#### 이제 특정 t 시점에서의 상태 s가 주어졌다는 상황에서 수학적 모델링을 표현해보고 모든 시점으로 확대해보자

### 목표 :
- 이제 에이전트와 환경 간의 수학적 모델링이 끝났으니
- 강화학습에서 에이전트의 목적을 다시 상기해보자

- 복습 :
    1. 상태 $s$ 가 주어졌을 때 다음 행동의 확률을 나타내는 정책 $\pi(a|s)$
    2. 상태 $s$ 와 행동 $a$ 가 주어졌을 때 다음 상태로 전이되는 확률 $p(s'|s,a)$
    3. 그리고 주어지는 보상 $r(s,a,s')$

- t 시점의 상태 $s$ 에서 미래까지 반영하여 높은 보상을 얻도록 에이전트의 행동을 결정하는 것이 목표임.
- 이러한 개념을 "수익" return 이라함.
- Return을 최대화하는 것이 에이전트의 목적임

시간 t 에서의 Return $G_t$ 는 다음과 같이 정의됨.
- $G_t=R_t + \gamma R_{t+1} + \gamma^2 R_{t+2} +...$ , $(0<\gamma<1)$ ,$\gamma$ : 할인율 

- 할인율을 반영한 주된 이유는 2가지로 추측됨.
    1. Return이 무한대로 발산하는 문제를 방지.
    2. 현재 시점에서의 보상을 더 중요하게 생각하는 것을 반영한다고 해석.

#### 
#### 이제 t 시점에서의 Return을 극대화하는 것이 에이전트의 목적으로 수학식으로 표현 가능
- 현재 논의에서는 극대화된 것을 찾는 것이 아니라 에이전트의 정책이 얼마나 좋고 나쁜지를 정량적 평가하는 측면에서 보자


1. 이때 중요한 것은 Return의 기댓값임.
2. t 시점에서 상태 s 가 결정적으로 주어졌지만, 
3. 에이전트의 정책(즉, 행동이 확률적으로 움직임) $\pi$ 는 확률적이므로
4. t 시점에서의 에이전트의 나타날 모든 행동들에 대응하는 Return의 모든 값들을 계산하여 
5. 평가할 수 있음.
6. 그래서 기댓값이 취해지는 함수로 표현됨.

$v_\pi(s) $ 함수를 상태 가치 함수(State Value function) 이라 함.

$\pi$ 가 확률을 나타내므로 아래와 같이도 표기  가능


#### 상태가치함수 $v_\pi(s)$ 해석 :

### 
강화학습 목표는 에이전트의 최적 정책을 찾는 것임.
- 모든 $\pi$ 들에 대해 $v_{\pi_{*}}(s) \geq v_\pi(s)$ 을 만족하는 정책 $\pi_{*}$ 을 최적 정책이라함

- 최적정책 optimal policy $\pi_{*}$ 의 상태 가치 함수를 최적 상태 가치 함수 optimal state value function $v_{\pi_{*}}(s)$ 라고 함

#### 정리 :
1. Markov Decision Process 는 강화학습을 수학적으로 모델링하는 하나의 방법을 의미함.(상태전이, 에이전트 정책, 보상 3가지 단계로 구분, 함수로 모델링)
2. 강화학습의 목적은 에이전트가 각 상태에서 미래까지 고려한 수익을 최대화하는 것을 목표로함.
3. 이러한 과정이 일반적으로 확률적이므로 기댓값 개념을 취해 사용

#### 
#### 에이전트의 행동이 결정적인 경우
L1인 상태에서 오른쪽으로만 가는 행동만 하는 에이전트를 방향있는 그래프로 표현해보면 다음과 같음

#### 에이전트의 행동이 확률적인 경우
L1인 상태에서 에이전트가 오른쪽, 왼쪽 50% 확률로 움직일 수 있다면 


# 
#### 지금의 논의에서는 에이전트가 행동을 결정적으로 하는 경우에 대해 들여다보겠음
#### 그리고 환경의 전이 상태도 결정적인 경우로 다뤄보자

결정적 정책 $a =\mu(s)$ , 상태 s 가 주어지면 왼쪽, 오른쪽 무조건 100% 가는 것

그렇다면 에이전트가 L1, L2 인 두 상태에서 취할 수 있는 정책의 모든 경우의 수는 아래와 같이 4가지임


1. 만약 상태가 L1 으로 시작하고, 정책을 $\mu_1(s)$ 을 따른다면 L1 상태에서 상태가치함수는 다음과 같이 계산됨


```python
V = 1 # L1 상태에서 오른쪽으로 이동하여 보상을 얻은 것(처음)

for i in range(1, 100):
    V += -1 * (0.9 ** i)
    # print("L1 상태에서 오른쪽으로 이동하여 보상을 얻은 것(반복) ", V)
print("L1 상태에서 시작하여 모든 상태에서 오른쪽으로만 행동하는 에이전트의 상태가치함수 값은 ", V)
```

2. 만약 L2 인 상태에서 시작하여 정책 $\mu_1(s)$ 을 따르는 상태가치함수는 다음과 같음


```python
V = -1 # L2 상태에서 오른쪽으로 이동하면 벽이므로 -1 보상(처음)

for i in range(1, 100):
    V += -1 * (0.9 ** i)
print("L2 상태에서 시작하여 모든 상태에서 오른쪽으로만 행동하는 에이전트의 상태가치함수 값은 ", V)
```

#### 문제 : 나머지 정책 3개에 대해서도 각 L1, L2 상태일 때에 상태가치함수 값을 계산하여 나타내보자


```python
a = 1
a % 2
```

```python
V = 0

for i in range(1, 100): # 이건 첫 리워드부터 가중치를 부여한거라 잘못된 건
    if i % 2 == 1:
        V += 1 * (0.9 ** i)
    else:
        V += 0 * (0.9 ** i)
print(f"\nL1 상태에서 시작하여 정책 μ_2를 따르는 에이전트의 상태가치함수 값: {V:.4f}")
```

```python
V = 1  # L1에서 오른쪽으로 이동하여 L2로 가면서 +1 보상 (첫 번째 행동)

# L1 → L2 → L1 → L2 → ... 패턴으로 무한 반복
# 홀수 번째 시점(i=1,3,5,...): L2에서 왼쪽으로 L1로 이동, 보상 0
# 짝수 번째 시점(i=2,4,6,...): L1에서 오른쪽으로 L2로 이동, 보상 +1

for i in range(1, 100):
    if i % 2 == 1:  # 홀수 번째: L2 → L1 (보상 0)
        V += 0 * (0.9 ** i)
    else:  # 짝수 번째: L1 → L2 (보상 +1)
        V += 1 * (0.9 ** i)

print(f"\nL1 상태에서 시작하여 정책 μ₂를 따르는 에이전트의 상태가치함수 값: {V:.4f}")
```

```python
V = 0 # L2 상태에서 왼쪽으로 이동하여 L1로 가면서 +1 보상 (첫 번째 행동)

for i in range(1, 100):
    if i % 2 == 0: # L2 -> L1 (보상 0)
        V += 0 * (0.9 ** i)
    else: # L1 -> L2 (보상 -1)
        V += 1 * (0.9 ** i)
print(f"L2 상태에서 시작하여 정책 μ₂를 따르는 에이전트의 상태가치함수 값: {V:.4f}")
```

```python
V = -1 # L1 상태에서 3번째 정책을 따를 때

for i in range(1, 100):
    V += -1 * (0.9 ** i)
print(V)
```

```python
V = -1 # L2 상태에서 3번째 정책을 따를 때

for i in range(1, 100):
    V += -1 * (0.9 ** i)
print(V)
```

- 2번째 정책가치함수

```python
# 2번째 정책과 L1 상태에서 상태가치함수
V = 1

for i in range(1, 100):
    if i % 2 == 1:
        V += 0
    else :
        V += 1 * (0.9 ** i)
print("L1 상태 시작, 2번째 정책 : ", V)
```

```python
# 2번째 정책과 L2 상태에서 상태가치함수
V = 0

for i in range(1, 100):
    if i % 2 == 1:
        V += 1 * (0.9 ** i)
    else :
        V += 0
print("L2 상태 시작, 2번째 정책 : ", V)
```
